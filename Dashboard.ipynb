## 6. Similar company:

```python
# Here finding 5 similar companies on the basis of similar Industries and same city
```

```python
# Function definition

def find_similar_companies_contains(company):
    target_row = df[df['Company'] == company]
    
    target_industry = str(target_row.iloc[0]['Industry']).strip().lower()
    target_city = str(target_row.iloc[0]['City']).strip().lower()

    df['Industry'] = df['Industry'].astype(str)
    df['City'] = df['City'].astype(str)

    similar = df[
        (df['Company'] != company) &
        (df['Industry'].str.lower().str.contains(target_industry, na=False)) &
        (df['City'].str.lower().str.contains(target_city, na=False))
    ]
    
    return similar['Company'].unique()[:5]
```

```python
# Function call

find_similar_companies_contains('Oyo')
```
